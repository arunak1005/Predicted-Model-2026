```python
# ============================================================
# 🔥 FULL PIPELINE: TUNED GNN with Multi-Code Embeddings + Imbalance Handling
# ============================================================

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from torch.nn.utils.rnn import pad_sequence
import matplotlib.pyplot as plt
import seaborn as sns

# ============================================================
# 1️⃣ COPY DATA & TARGET
# ============================================================

df = df_new.copy()
df['target'] = df['InitialDeterminationStatus'].map({"APPROVED":0, "DENIED":1})

# ============================================================
# 2️⃣ OUTLIER CAPPING
# ============================================================

num_cols = ["Payments", "LOS", "AGE"]
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    df[col] = df[col].clip(Q1 - 1.5*IQR, Q3 + 1.5*IQR)

# ============================================================
# 3️⃣ LOG TRANSFORM
# ============================================================

df["Payments_log"] = np.log1p(df["Payments"])
df["LOS_log"] = np.log1p(df["LOS"])

# ============================================================
# 4️⃣ TRAIN-TEST SPLIT
# ============================================================

train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['target'])

# ============================================================
# 5️⃣ TARGET ENCODING
# ============================================================

def target_encode(train, test, col):
    mapping = train.groupby(col)['target'].mean()
    train[col + "_enc"] = train[col].map(mapping)
    test[col + "_enc"] = test[col].map(mapping)
    global_mean = train['target'].mean()
    test[col + "_enc"] = test[col + "_enc"].fillna(global_mean)
    return train, test

train_df, test_df = target_encode(train_df, test_df, "ADRG")
train_df, test_df = target_encode(train_df, test_df, "PRIM_DX")

# ============================================================
# 6️⃣ MULTI-CODE FEATURES → EMBEDDING INDICES
# ============================================================

multi_cols = ["A_DX_MCC_Set", "A_DX_CC_Set", "A_DX_General_Set", "A_PX_List"]
encoders = {}
for col in multi_cols:
    all_codes = set()
    for codes in train_df[col].dropna():
        all_codes.update([c.strip() for c in codes.split(',')])
    enc = {c:i+1 for i,c in enumerate(all_codes)}
    encoders[col] = enc

def encode_multi(df, col, enc):
    def encode_row(row):
        if pd.isna(row):
            return [0]
        return [enc.get(c.strip(), 0) for c in row.split(',')]
    return df[col].apply(encode_row)

for col in multi_cols:
    train_df[col + "_idx"] = encode_multi(train_df, col, encoders[col])
    test_df[col + "_idx"] = encode_multi(test_df, col, encoders[col])

# ============================================================
# 7️⃣ NUMERIC FEATURES SCALING
# ============================================================

numeric_features = ["Payments", "Payments_log", "LOS", "LOS_log", "AGE", "ADRG_enc", "PRIM_DX_enc"]
scaler = RobustScaler()
train_df[numeric_features] = scaler.fit_transform(train_df[numeric_features])
test_df[numeric_features] = scaler.transform(test_df[numeric_features])

# ============================================================
# 8️⃣ PYTORCH DATASET
# ============================================================

class PatientDataset(Dataset):
    def __init__(self, df, num_cols, code_cols):
        self.num_features = torch.tensor(df[num_cols].values, dtype=torch.float32)
        self.code_features = [df[c+"_idx"].tolist() for c in code_cols]
        self.y = torch.tensor(df['target'].values, dtype=torch.float32)
        
    def __len__(self):
        return len(self.y)
    
    def __getitem__(self, idx):
        return self.num_features[idx], [c[idx] for c in self.code_features], self.y[idx]

train_dataset = PatientDataset(train_df, numeric_features, multi_cols)
test_dataset = PatientDataset(test_df, numeric_features, multi_cols)

# Weighted sampler for imbalance
class_counts = train_df['target'].value_counts().sort_index()
weights = 1.0 / class_counts
sample_weights = train_df['target'].map(weights).values
sampler = WeightedRandomSampler(sample_weights, num_samples=len(sample_weights), replacement=True)

def collate_fn(batch):
    num_feats = torch.stack([b[0] for b in batch])
    code_feats = []
    for i in range(len(multi_cols)):
        code_list = [torch.tensor(b[1][i], dtype=torch.long) for b in batch]
        code_feats.append(code_list)
    labels = torch.tensor([b[2] for b in batch], dtype=torch.float32).unsqueeze(1)
    return num_feats, code_feats, labels

train_loader = DataLoader(train_dataset, batch_size=64, sampler=sampler, collate_fn=collate_fn)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False, collate_fn=collate_fn)

# ============================================================
# 9️⃣ TUNED GNN MODEL
# ============================================================

class GNNModel(nn.Module):
    def __init__(self, numeric_dim, code_cardinalities, embed_dim=32, hidden_dim=128, dropout=0.3):
        super().__init__()
        self.embeddings = nn.ModuleList([nn.Embedding(card+1, embed_dim) for card in code_cardinalities])
        self.fc1 = nn.Linear(numeric_dim + embed_dim*len(code_cardinalities), hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        self.dropout = nn.Dropout(dropout)
        self.out = nn.Linear(hidden_dim, 1)
        
    def forward(self, x_num, x_code_list):
        embed_list = []
        for emb, codes in zip(self.embeddings, x_code_list):
            padded = pad_sequence(codes, batch_first=True)
            emb_out = emb(padded)
            mean_emb = emb_out.mean(dim=1)
            embed_list.append(mean_emb)
        x = torch.cat([x_num] + embed_list, dim=1)
        x = F.relu(self.fc1(x))
        x = self.dropout(F.relu(self.fc2(x)))
        return self.out(x)  # logits

code_cardinalities = [len(encoders[c]) for c in multi_cols]
model = GNNModel(numeric_dim=len(numeric_features), code_cardinalities=code_cardinalities)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# ============================================================
# 🔟 BCEWithLogitsLoss WITH POS_WEIGHT
# ============================================================

pos_weight = torch.tensor(class_counts[0]/class_counts[1], dtype=torch.float32).to(device)
criterion = nn.BCEWithLogitsLoss(pos_weight=pos_weight)
optimizer = torch.optim.Adam(model.parameters(), lr=5e-4)

# ============================================================
# 1️⃣1️⃣ TRAINING LOOP
# ============================================================

num_epochs = 50
for epoch in range(1, num_epochs+1):
    model.train()
    epoch_loss = 0
    for x_num_batch, x_code_batch, y_batch in train_loader:
        x_num_batch = x_num_batch.to(device)
        x_code_batch = [pad_sequence(codes, batch_first=True).to(device) for codes in x_code_batch]
        y_batch = y_batch.to(device)
        
        optimizer.zero_grad()
        logits = model(x_num_batch, x_code_batch)
        loss = criterion(logits, y_batch)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    if epoch % 10 == 0:
        print(f"Epoch {epoch}/{num_epochs}, Loss: {epoch_loss/len(train_loader):.4f}")

# ============================================================
# 1️⃣2️⃣ EVALUATION
# ============================================================

model.eval()
y_true, y_pred_prob = [], []
with torch.no_grad():
    for x_num_batch, x_code_batch, y_batch in test_loader:
        x_num_batch = x_num_batch.to(device)
        x_code_batch = [pad_sequence(codes, batch_first=True).to(device) for codes in x_code_batch]
        y_batch = y_batch.to(device)
        logits = model(x_num_batch, x_code_batch)
        y_true.extend(y_batch.cpu().numpy())
        y_pred_prob.extend(torch.sigmoid(logits).cpu().numpy())

y_pred_class = (np.array(y_pred_prob) > 0.5).astype(int)
print(classification_report(y_true, y_pred_class))
print("ROC-AUC:", roc_auc_score(y_true, y_pred_prob))

cm = confusion_matrix(y_true, y_pred_class)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
import pandas as pd
import torch

# Suppose your GNNModel is called 'model' and is already trained
# multi_cols = ["A_DX_MCC_Set", "A_DX_CC_Set", "A_DX_General_Set", "A_PX_List"]
# encoders = dictionary of code -> index for each code type

def get_embeddings_df(model, encoders, multi_cols):
    embedding_dfs = {}
    
    for col, enc in zip(multi_cols, encoders.values()):
        # Access embedding layer corresponding to this code type
        emb_layer = model.embeddings[list(encoders.keys()).index(col)]
        
        # Get embedding weights (tensor) -> convert to numpy
        emb_weights = emb_layer.weight.detach().cpu().numpy()
        
        # Map indices back to code names
        idx_to_code = {v: k for k, v in enc.items()}
        
        # Build DataFrame
        df_emb = pd.DataFrame(emb_weights, index=[idx_to_code.get(i, 'PAD') for i in range(emb_weights.shape[0])])
        df_emb.index.name = col
        df_emb.columns = [f"dim_{i+1}" for i in range(emb_weights.shape[1])]
        
        embedding_dfs[col] = df_emb
        
    return embedding_dfs

# Example usage:
embedding_dfs = get_embeddings_df(model, encoders, multi_cols)

# To see MCC embeddings:
#print(embedding_dfs['A_DX_MCC_Set'])

# To export to Excel if you want:
with pd.ExcelWriter("embeddings.xlsx") as writer:
    for col, df_emb in embedding_dfs.items():
        df_emb.to_excel(writer, sheet_name=col)
```

```python
# ============================================================
# 🔥 FULL PIPELINE: GNN with Multi-Code Embeddings + Imbalance Handling
# ============================================================

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.nn.utils.rnn import pad_sequence
import matplotlib.pyplot as plt
import seaborn as sns

# ============================================================
# 1️⃣ COPY DATA & TARGET
# ============================================================

df = df_new.copy()
df['target'] = df['InitialDeterminationStatus'].map({"APPROVED":0, "DENIED":1})

# ============================================================
# 2️⃣ OUTLIER CAPPING
# ============================================================

num_cols = ["Payments", "LOS", "AGE"]
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    df[col] = df[col].clip(Q1 - 1.5*IQR, Q3 + 1.5*IQR)

# ============================================================
# 3️⃣ LOG TRANSFORM
# ============================================================

df["Payments_log"] = np.log1p(df["Payments"])
df["LOS_log"] = np.log1p(df["LOS"])

# ============================================================
# 4️⃣ TRAIN-TEST SPLIT
# ============================================================

train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['target'])

# ============================================================
# 5️⃣ TARGET ENCODING FOR LOW-CARDINALITY FEATURES
# ============================================================

def target_encode(train, test, col):
    mapping = train.groupby(col)['target'].mean()
    train[col + "_enc"] = train[col].map(mapping)
    test[col + "_enc"] = test[col].map(mapping)
    global_mean = train['target'].mean()
    test[col + "_enc"] = test[col + "_enc"].fillna(global_mean)
    return train, test

train_df, test_df = target_encode(train_df, test_df, "ADRG")
train_df, test_df = target_encode(train_df, test_df, "PRIM_DX")

# ============================================================
# 6️⃣ MULTI-CODE FEATURES → EMBEDDING INDICES
# ============================================================

multi_cols = ["A_DX_MCC_Set", "A_DX_CC_Set", "A_DX_General_Set", "A_PX_List"]
encoders = {}
for col in multi_cols:
    all_codes = set()
    for codes in train_df[col].dropna():
        all_codes.update([c.strip() for c in codes.split(',')])
    enc = {c:i+1 for i,c in enumerate(all_codes)}  # reserve 0 for padding
    encoders[col] = enc

def encode_multi(df, col, enc):
    def encode_row(row):
        if pd.isna(row):
            return [0]
        return [enc.get(c.strip(), 0) for c in row.split(',')]
    return df[col].apply(encode_row)

for col in multi_cols:
    train_df[col + "_idx"] = encode_multi(train_df, col, encoders[col])
    test_df[col + "_idx"] = encode_multi(test_df, col, encoders[col])

# ============================================================
# 7️⃣ NUMERIC FEATURES SCALING
# ============================================================

numeric_features = ["Payments", "Payments_log", "LOS", "LOS_log", "AGE", "ADRG_enc", "PRIM_DX_enc"]
scaler = RobustScaler()
train_df[numeric_features] = scaler.fit_transform(train_df[numeric_features])
test_df[numeric_features] = scaler.transform(test_df[numeric_features])

# ============================================================
# 8️⃣ PYTORCH DATASET
# ============================================================

class PatientDataset(Dataset):
    def __init__(self, df, num_cols, code_cols):
        self.num_features = torch.tensor(df[num_cols].values, dtype=torch.float32)
        self.code_features = [df[c+"_idx"].tolist() for c in code_cols]  # keep as list of lists
        self.y = torch.tensor(df['target'].values, dtype=torch.float32)
        
    def __len__(self):
        return len(self.y)
    
    def __getitem__(self, idx):
        return self.num_features[idx], [c[idx] for c in self.code_features], self.y[idx]

train_dataset = PatientDataset(train_df, numeric_features, multi_cols)
test_dataset = PatientDataset(test_df, numeric_features, multi_cols)

# Custom collate_fn to pad code features
def collate_fn(batch):
    num_feats = torch.stack([b[0] for b in batch])
    code_feats = []
    for i in range(len(multi_cols)):
        code_list = [torch.tensor(b[1][i], dtype=torch.long) for b in batch]
        code_feats.append(code_list)
    labels = torch.tensor([b[2] for b in batch], dtype=torch.float32).unsqueeze(1)
    return num_feats, code_feats, labels

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True, collate_fn=collate_fn)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False, collate_fn=collate_fn)

# ============================================================
# 9️⃣ GNN MODEL WITH EMBEDDINGS
# ============================================================

class GNNModel(nn.Module):
    def __init__(self, numeric_dim, code_cardinalities, embed_dim=16, hidden_dim=64):
        super().__init__()
        self.embeddings = nn.ModuleList([nn.Embedding(card+1, embed_dim) for card in code_cardinalities])
        total_input = numeric_dim + embed_dim * len(code_cardinalities)
        self.fc1 = nn.Linear(total_input, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        self.out = nn.Linear(hidden_dim, 1)
        
    def forward(self, x_num, x_code_list):
        embed_list = []
        for emb, codes in zip(self.embeddings, x_code_list):
            padded = pad_sequence(codes, batch_first=True)  # [batch, max_len]
            emb_out = emb(padded)  # [batch, max_len, embed_dim]
            mean_emb = emb_out.mean(dim=1)  # [batch, embed_dim]
            embed_list.append(mean_emb)
        x = torch.cat([x_num] + embed_list, dim=1)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        return torch.sigmoid(self.out(x))

code_cardinalities = [len(encoders[c]) for c in multi_cols]
model = GNNModel(numeric_dim=len(numeric_features), code_cardinalities=code_cardinalities)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# ============================================================
# 🔟 LOSS & OPTIMIZER WITH IMBALANCE HANDLING
# ============================================================

pos_weight = torch.tensor((train_df['target']==0).sum() / (train_df['target']==1).sum(), dtype=torch.float32).to(device)
criterion = nn.BCELoss()  # you can also implement weighted loss using pos_weight
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# ============================================================
# 1️⃣1️⃣ TRAINING LOOP
# ============================================================

num_epochs = 50
for epoch in range(1, num_epochs+1):
    model.train()
    epoch_loss = 0
    for x_num_batch, x_code_batch, y_batch in train_loader:
        x_num_batch = x_num_batch.to(device)
        x_code_batch = [pad_sequence(codes, batch_first=True).to(device) for codes in x_code_batch]
        y_batch = y_batch.to(device)
        
        optimizer.zero_grad()
        y_pred = model(x_num_batch, x_code_batch)
        loss = criterion(y_pred, y_batch)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    if epoch % 10 == 0:
        print(f"Epoch {epoch}/{num_epochs}, Loss: {epoch_loss/len(train_loader):.4f}")

# ============================================================
# 1️⃣2️⃣ EVALUATION
# ============================================================

model.eval()
y_true, y_pred_prob = [], []
with torch.no_grad():
    for x_num_batch, x_code_batch, y_batch in test_loader:
        x_num_batch = x_num_batch.to(device)
        x_code_batch = [pad_sequence(codes, batch_first=True).to(device) for codes in x_code_batch]
        y_batch = y_batch.to(device)
        y_pred = model(x_num_batch, x_code_batch)
        y_true.extend(y_batch.cpu().numpy())
        y_pred_prob.extend(y_pred.cpu().numpy())

y_pred_class = (np.array(y_pred_prob) > 0.5).astype(int)
print(classification_report(y_true, y_pred_class))
print("ROC-AUC:", roc_auc_score(y_true, y_pred_prob))

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred_class)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# ============================================================
# 🔥 1. SETUP
# ============================================================

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.ensemble import StackingClassifier
from sklearn.base import BaseEstimator, ClassifierMixin
import seaborn as sns
import matplotlib.pyplot as plt

from xgboost import XGBClassifier
import lightgbm as lgb
from imblearn.over_sampling import RandomOverSampler

df = df_new.copy()

# Target
df['target'] = df['InitialDeterminationStatus'].map({
    "APPROVED": 0,
    "DENIED": 1
})

# ============================================================
# ✂️ 2. OUTLIER CAPPING
# ============================================================

num_cols = ["Payments", "LOS", "AGE"]
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    df[col] = df[col].clip(lower, upper)

# ============================================================
# 🔄 3. LOG TRANSFORM
# ============================================================

df["Payments_log"] = np.log1p(df["Payments"])
df["LOS_log"] = np.log1p(df["LOS"])

# ============================================================
# 🎯 4. TRAIN-TEST SPLIT
# ============================================================

train_df, test_df = train_test_split(
    df, test_size=0.2, random_state=42, stratify=df['target']
)

# ============================================================
# 🧠 5. TARGET ENCODING
# ============================================================

def target_encode(train, test, col):
    mapping = train.groupby(col)['target'].mean()
    global_mean = train['target'].mean()
    train[col + "_enc"] = train[col].map(mapping)
    test[col + "_enc"] = test[col].map(mapping)
    test[col + "_enc"] = test[col + "_enc"].fillna(global_mean)
    return train, test

for col in ["ADRG","PRIM_DX"]:
    train_df, test_df = target_encode(train_df,test_df,col)

# ============================================================
# 🧠 6. STATISTICAL EMBEDDINGS
# ============================================================

def explode_column(df, col):
    temp = df[[col,'target']].dropna()
    temp[col] = temp[col].str.split(',')
    temp = temp.explode(col)
    temp[col] = temp[col].str.strip()
    return temp

def create_code_stats(df, col):
    temp = explode_column(df,col)
    stats = temp.groupby(col).agg({'target':['mean','count']})
    stats.columns = ['denial_rate','freq']
    return stats.reset_index()

def aggregate_features(df, col, stats, prefix):
    stats_dict = stats.set_index(col).to_dict(orient='index')
    def compute(row):
        if pd.isna(row):
            return [0,0]
        codes = str(row).split(',')
        denial_rates,freqs=[],[]
        for code in codes:
            code = code.strip()
            if code in stats_dict:
                denial_rates.append(stats_dict[code]['denial_rate'])
                freqs.append(stats_dict[code]['freq'])
        if len(denial_rates)==0:
            return [0,0]
        return [np.mean(denial_rates), np.mean(freqs)]
    features = df[col].apply(compute)
    df[f'{prefix}_denial_mean'] = features.apply(lambda x:x[0])
    df[f'{prefix}_freq_mean'] = features.apply(lambda x:x[1])
    return df

cols = ["A_DX_MCC_Set","A_DX_CC_Set","A_DX_General_Set","A_PX_List"]
prefixes = ["MCC","CC","GEN","PX"]
stats_dict = {col:create_code_stats(train_df,col) for col in cols}

for col,prefix in zip(cols,prefixes):
    train_df = aggregate_features(train_df,col,stats_dict[col],prefix)
    test_df = aggregate_features(test_df,col,stats_dict[col],prefix)

# ============================================================
# 🎯 7. FINAL FEATURES
# ============================================================

features = [
    "Payments","Payments_log","LOS","LOS_log","AGE",
    "ADRG_enc","PRIM_DX_enc",
    "MCC_denial_mean","MCC_freq_mean",
    "CC_denial_mean","CC_freq_mean",
    "GEN_denial_mean","GEN_freq_mean",
    "PX_denial_mean","PX_freq_mean"
]

X_train = train_df[features].values
X_test = test_df[features].values
y_train = train_df["target"].values
y_test = test_df["target"].values

# ============================================================
# ⚙️ 8. SCALE & BALANCE
# ============================================================

scaler = RobustScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Oversample minority class for NN
ros = RandomOverSampler(random_state=42)
X_train_res, y_train_res = ros.fit_resample(X_train, y_train)

# ============================================================
# 🧠 9. PYTORCH NN WRAPPER FOR STACKING
# ============================================================
class TorchNN(BaseEstimator, ClassifierMixin):
    def __init__(self, input_dim, epochs=50, lr=0.001):
        self.input_dim = input_dim
        self.epochs = epochs
        self.lr = lr
        self.device = "cuda" if torch.cuda.is_available() else "cpu"
        self._build_model()
    
    def _build_model(self):
        self.model = nn.Sequential(
            nn.Linear(self.input_dim,64),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(64,32),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(32,1)
        ).to(self.device)
    
    def fit(self, X, y):
        # Save classes for sklearn compatibility
        self.classes_ = np.unique(y)
        y_t = torch.tensor(y.reshape(-1,1), dtype=torch.float32).to(self.device)
        X_t = torch.tensor(X, dtype=torch.float32).to(self.device)
        
        # Compute pos_weight for imbalance
        pos_weight = torch.tensor([(y==0).sum() / (y==1).sum()], dtype=torch.float32).to(self.device)
        criterion = nn.BCEWithLogitsLoss(pos_weight=pos_weight)
        optimizer = torch.optim.Adam(self.model.parameters(), lr=self.lr)
        
        self.model.train()
        for epoch in range(self.epochs):
            optimizer.zero_grad()
            output = self.model(X_t)
            loss = criterion(output, y_t)
            loss.backward()
            optimizer.step()
        return self
    
    def predict_proba(self, X):
        X_t = torch.tensor(X, dtype=torch.float32).to(self.device)
        self.model.eval()
        with torch.no_grad():
            logits = self.model(X_t)
            probs = torch.sigmoid(logits).cpu().numpy()
        return np.hstack([1-probs, probs])
    
    def predict(self, X):
        return (self.predict_proba(X)[:,1] >= 0.5).astype(int)

# ============================================================
# 🔥 10. STACKING MODEL
# ============================================================

nn_clf = TorchNN(input_dim=X_train.shape[1], epochs=50, lr=0.001)
xgb_clf = XGBClassifier(
    n_estimators=300, max_depth=6, learning_rate=0.05,
    subsample=0.8, colsample_bytree=0.8, scale_pos_weight=(y_train==0).sum()/(y_train==1).sum(),
    random_state=42, eval_metric='logloss'
)
lgb_clf = lgb.LGBMClassifier(
    n_estimators=300, max_depth=6, learning_rate=0.05,
    subsample=0.8, colsample_bytree=0.8,
    class_weight='balanced', random_state=42
)

stack = StackingClassifier(
    estimators=[('nn', nn_clf), ('xgb', xgb_clf), ('lgb', lgb_clf)],
    final_estimator=XGBClassifier(n_estimators=200, max_depth=4, learning_rate=0.05, random_state=42),
    n_jobs=-1, passthrough=True
)

stack.fit(X_train, y_train)

# ============================================================
# 📊 11. EVALUATION
# ============================================================

y_pred = stack.predict(X_test)
y_prob = stack.predict_proba(X_test)[:,1]

print(classification_report(y_test,y_pred))
print("ROC-AUC:", roc_auc_score(y_test,y_prob))

cm = confusion_matrix(y_test,y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Stacked Model Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
def predict_claim_full(input_dict, model, scaler, stats_dict, features, train_df):
    
    import pandas as pd
    import numpy as np
    
    df_input = pd.DataFrame([input_dict])
    
    # ======================================================
    # 🔄 1. LOG TRANSFORM
    # ======================================================
    df_input["Payments_log"] = np.log1p(df_input["Payments"])
    df_input["LOS_log"] = np.log1p(df_input["LOS"])
    
    # ======================================================
    # 🧠 2. TARGET ENCODING (ADRG, PRIM_DX)
    # ======================================================

    for col in ["ADRG", "PRIM_DX"]:
        mapping = train_df.groupby(col)['target'].mean()
        global_mean = train_df['target'].mean()
        
        df_input[col + "_enc"] = df_input[col].map(mapping)
        
        # FIXED LINE ✅
        df_input[col + "_enc"] = df_input[col + "_enc"].fillna(global_mean)
  
    
    # ======================================================
    # 🧠 3. STATISTICAL EMBEDDING
    # ======================================================
    def apply_embedding(row, col, stats):
        stats_map = stats.set_index(col).to_dict(orient='index')
        
        if pd.isna(row[col]):
            return 0, 0
        
        codes = str(row[col]).split(',')
        
        denial_rates, freqs = [], []
        
        for code in codes:
            code = code.strip()
            if code in stats_map:
                denial_rates.append(stats_map[code]['denial_rate'])
                freqs.append(stats_map[code]['freq'])
        
        if len(denial_rates) == 0:
            return 0, 0
        
        return np.mean(denial_rates), np.mean(freqs)
    
    cols = ["A_DX_MCC_Set", "A_DX_CC_Set", "A_DX_General_Set", "A_PX_List"]
    prefixes = ["MCC", "CC", "GEN", "PX"]
    
    for col, prefix in zip(cols, prefixes):
        d_mean, f_mean = apply_embedding(df_input.iloc[0], col, stats_dict[col])
        
        df_input[f"{prefix}_denial_mean"] = d_mean
        df_input[f"{prefix}_freq_mean"] = f_mean
    
    # ======================================================
    # 🎯 4. SELECT FEATURES
    # ======================================================
    X = df_input[features]
    
    # ======================================================
    # ⚙️ 5. SCALE
    # ======================================================
    X_scaled = scaler.transform(X)
    
    # ======================================================
    # 🚀 6. PREDICTION
    # ======================================================
    prob = model.predict_proba(X_scaled)[0][1]
    pred = model.predict(X_scaled)[0]
    
    label = "DENIED" if pred == 1 else "APPROVED"
    
    return {
        "Prediction": label,
        "Denied_Confidence_%": round(prob * 100, 2),
        "Approved_Confidence_%": round((1 - prob) * 100, 2)
    }
```

```python
df_pred_copy = df_pred.copy()

results = []

for _, row in df_pred_copy.iterrows():
    
    input_dict = row.to_dict()
    
    output = predict_claim_full(
        input_dict=input_dict,
        model=model,
        scaler=scaler,
        stats_dict=stats_dict,
        features=features,
        train_df=train_df
    )
    
    results.append(output)

# Convert results to DataFrame
df_results = pd.DataFrame(results)

# Merge with original
df_pred_final = pd.concat([df_pred_copy.reset_index(drop=True), df_results], axis=1)

df_pred_final.shape
```
